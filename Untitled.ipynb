```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
from sklearn.cluster import AgglomerativeClustering
import datetime as dt
```

```python
kupci = pd.read_csv('OnlineRetail.csv')
df_null = round(100*(kupci.isnull().sum())/len(kupci), 2)
df_null
```

```python
kupci = kupci.dropna()
kupci.shape
```

```python
kupci.describe()
```

```python
kupci['CustomerID'] = kupci['CustomerID'].astype(str)
```

```python
kupci['UkupnaCijena'] = kupci['Quantity']*kupci['UnitPrice']
rfm_m = kupci.groupby('CustomerID')['UkupnaCijena'].sum()
rfm_m = rfm_m.reset_index()
rfm_m.head()
```

```python
rfm_f = kupci.groupby('CustomerID')['InvoiceNo'].count()
rfm_f = rfm_f.reset_index()
rfm_f.columns = ['CustomerID', 'BrojTransakcija']
rfm_f.head()
```

```python
rfm = pd.merge(rfm_m, rfm_f, on='CustomerID', how='inner')
rfm.head()
```

```python
attributes = ['UkupnaCijena','BrojTransakcija']
plt.rcParams['figure.figsize'] = [10,8]
sns.boxplot(data = rfm[attributes], orient="v", palette="Set2" ,whis=1.5,saturation=1, width=0.7)
plt.title("Distribucija promjenjivih", fontsize = 14, fontweight = 'bold')
plt.ylabel("Raspon", fontweight = 'bold')
plt.xlabel("Atributi", fontweight = 'bold')
```

```python
# Rescaling the attributes

rfm_df = rfm[['UkupnaCijena', 'BrojTransakcija']]

# Instantiate
scaler = StandardScaler()

# fit_transform
rfm_df_scaled = scaler.fit_transform(rfm_df)
rfm_df_scaled.shape
```

```python
mergings = linkage(rfm_df, method="average", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
mergings = linkage(rfm_df, method="complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
mergings = linkage(rfm_df, method="single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
cluster_labels = cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
rfm['Cluster_Labels'] = cluster_labels
rfm.head()
```

```python
sns.boxplot(x='Cluster_Labels', y='UkupnaCijena', data=rfm)
```

```python
sns.boxplot(x='Cluster_Labels', y='BrojTransakcija', data=rfm)
```

```python
X = rfm.iloc[:, [1, 2]].values
X
rfm
```

```python
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('Klasteri kupaca')
plt.xlabel('UkupnaCijena')
plt.ylabel('Broj transakcija')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,10))
dendrogram = dendrogram(linkage(X, method = 'ward'))
plt.title('Dendrogram', color='white')
plt.xlabel('Kupci', color='white')
plt.ylabel('Euclidova distanca', color='white')
plt.show()
```
